```python
x=ds_new.drop([13,1,7,6,5],axis=1)
print(x.shape)
y=ds_new[13]
print(y.shape)
```

separate input and output data for classification models

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

Scaled input data for model

```python
def maxr2_score(rg,x,y):
    max_r_score =0
    for r_state in range(42,200):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        rg.fit(x_train,y_train)
        pred=rg.predict(x_test)
        a_scr =accuracy_score(y_test,pred)
        if a_scr>max_r_score:
            max_r_score=a_scr
            final_r_state=r_state
    print('max r2 score corresponding',final_r_state,'is',max_r_score)
    return final_r_state
```

Created function for finding the accuracy score of model

# Applying different algorithms to create heartdisease classification model to find out the presence of heart disease in the patient

```python
lg=LogisticRegression()
r_state=maxr2_score(lg,x,y)
print('Mean r2 score for Logistic Regression:',cross_val_score(lg,x,y,cv=5).mean())
print('stander deviation in r2 score for Logistic Regression',cross_val_score(lg,x,y,cv=5,).std())
```

```python
models=['linear', 'poly', 'rbf', 'sigmoid']
```

```python
for i in models:
    svc=SVC(kernel=i)
    print('*******************************',i,'************************************************')
    print('\n')
    r_state=maxr2_score(svc,x,y)
    print('Mean r2 score for SVC:',cross_val_score(svc,x,y,cv=5).mean())
    print('stander deviation in r2 score for SVC',cross_val_score(svc,x,y,cv=5,).std())
    
```

```python
dtc=DecisionTreeClassifier()
r_state=maxr2_score(dtc,x,y)
print('Mean r2 score for Decision Tree Classifier:',cross_val_score(dtc,x,y,cv=5).mean())
print('stander deviation in r2 score for Decision Tree Classifier',cross_val_score(dtc,x,y,cv=5,).std())
    
```

```python
neighbors={'n_neighbors':range(1,30)}
knr=KNeighborsClassifier()
gknr=GridSearchCV(knr,neighbors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=28)
r_state=maxr2_score(knn,x,y)
print('Mean r2 score for KNeighbors Classifier:',cross_val_score(knn,x,y,cv=5).mean())
print('stander deviation in r2 score for KNeighbors Classifier:',cross_val_score(knn,x,y,cv=5,).std())
```

```python
ada_reg =AdaBoostClassifier()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000],"base_estimator":[lg,dtc]}
clf=GridSearchCV(ada_reg,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
ada_reg=AdaBoostClassifier(base_estimator=lg,learning_rate=0.001,n_estimators=1000)
r_state=maxr2_score(ada_reg,x,y)
print('Mean r2 score for Ada Boost Classifier:',cross_val_score(ada_reg,x,y,cv=5).mean())
print('stander deviation in r2 score for Ada Boost Classifier',cross_val_score(ada_reg,x,y,cv=5,).std())
```

```python
gb=GradientBoostingClassifier()
r_state=maxr2_score(gb,x,y)
print('Mean r2 score for Gradient Boosting:',cross_val_score(gb,x,y,cv=5).mean())
print('stander deviation in r2 score for Gradient Boosting',cross_val_score(gb,x,y,cv=5,).std())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=157,test_size=0.20)    
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
lg_pred=knn.predict(x_test)
print(accuracy_score(y_test,lg_pred))
print(confusion_matrix(y_test,lg_pred))
print(classification_report(y_test,lg_pred))
cvsdt=cross_val_score(knn,x,y,cv=10,scoring='accuracy').mean()
print('\n')
print(cvsdt)
print('\n')
```
