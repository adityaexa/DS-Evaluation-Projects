```python
ds_new['Active']=np.sqrt(ds_new['Active'])
ds_new['People_Tested']=np.sqrt(ds_new['People_Tested'])
ds_new['Confirmed']=np.sqrt(ds_new['Confirmed'])
ds_new['Deaths']=np.sqrt(ds_new['Deaths'])
ds_new['People_Hospitalized']=np.sqrt(ds_new['People_Hospitalized'])    
ds_new.skew()
```

Handling skewness of data by using square root method

```python
x_r=ds_new.drop(['Recovered','Incident_Rate','Mortality_Rate','Testing_Rate','Hospitalization_Rate','People_Hospitalized'],axis=1)
x_d=ds_new.drop(['Deaths','Incident_Rate','Mortality_Rate','Testing_Rate','Hospitalization_Rate','People_Hospitalized'],axis=1)
print(x_r.shape)
print(x_d.shape)
y_r=ds_new['Recovered']
y_d=ds_new['Deaths']
print(y_r.shape)
print(y_d.shape)
```

separate input and output data covid-19 for deaths(x_d,y_d) and recovered(x_r,y_r) cases.

```python
sc=StandardScaler()
x_r=sc.fit_transform(x_r)
x_r
```

Scaled input data for model

```python
def maxr2_score(rg,x,y):
    max_r_score =0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        rg.fit(x_train,y_train)
        pred=rg.predict(x_test)
        r2_scr=r2_score(y_test,pred)
       # print('r2 score corrasponding to',r_state,'is',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print('max r2 score corrasponding to ',final_r_state,'is',max_r_score)
    return final_r_state
```

Created function for finding the max r2 score value of model

# Applying different algorithms to create covid-19 Recovered model

```python
#Linear Regression model
lg_r=LinearRegression()
r_state=maxr2_score(lg_r,x_r,y_r)
print('Mean r2 score for linear Regression:',cross_val_score(lg_r,x_r,y_r,cv=5,scoring='r2').mean())
print('stander deviation in r2 score for linear Regression',cross_val_score(lg_r,x_r,y_r,cv=5,scoring='r2').std())
```

```python
neighbors={'n_neighbors':range(1,30)}
knr_r=KNeighborsRegressor()
gknr=GridSearchCV(knr_r,neighbors,cv=10)
gknr.fit(x_r,y_r)
gknr.best_params_
```

```python
knr_r=KNeighborsRegressor(n_neighbors=3)
r_state=maxr2_score(knr_r,x_r,y_r)
print('Mean r2 score for KNN Regression:',cross_val_score(knr_r,x_r,y_r,cv=5,scoring='r2').mean())
print('stander deviation r2 score for KNN Regression:',cross_val_score(knr_r,x_r,y_r,cv=5,scoring='r2').std())
```

```python
lsreg_r=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
clf=GridSearchCV(lsreg_r,parameters,cv=10)
clf.fit(x_r,y_r)
clf.best_params_
```

```python
#let check max r2 score when we use lasso
lsreg_r=Lasso(alpha=1)
r_state=maxr2_score(lsreg_r,x_r,y_r)
print('Mean r2 score for Lasso Regration',cross_val_score(lsreg_r,x_r,y_r,cv=5,scoring='r2').mean())
print('stander deviation in r2 score for Lasso Regretion',cross_val_score(lsreg_r,x_r,y_r,cv=5,scoring='r2').std())
```

```python
ada_reg_r =AdaBoostRegressor()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000],"base_estimator":[lg,lsreg,DecisionTreeRegressor()]}
clf=GridSearchCV(ada_reg_r,parameters,cv=5)
clf.fit(x_r,y_r)
clf.best_params_
```

```python
ada_reg_r=AdaBoostRegressor(base_estimator=lsreg,learning_rate=0.01,n_estimators=100)
r_state=maxr2_score(ada_reg_r,x_r,y_r)
print('Mean r2 score for ada Boosting Regration:',cross_val_score(ada_reg_r,x_r,y_r,cv=5,scoring='r2').mean())
print('Mean r2 score for ada Boosting Regration:',cross_val_score(ada_reg_r,x_r,y_r,cv=5,scoring='r2').std())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_r,y_r,random_state=96,test_size=0.20)
ada_reg_r.fit(x_train,y_train)
y_pred=ada_reg_r.predict(x_test)
print('RMSE is: ',np.sqrt(mean_squared_error(y_test,y_pred)))
print('MAE is: ',mean_absolute_error(y_test,y_pred))
print('r2_score is:',r2_score(y_test,y_pred))
print('cross val score Mean:',cross_val_score(ada_reg_r,x_r,y_r,cv=5,scoring='r2').mean())
print('cross val score std:',cross_val_score(ada_reg_r,x_r,y_r,cv=5,scoring='r2').std())
print('\n')
print(y_pred)
print(y_test)
```

Ada boosting regration model is slightly better than other models because it has less mean absolute error.

```python
joblib.dump(ada_reg_r,'covid_19_recovered.obj')
```
