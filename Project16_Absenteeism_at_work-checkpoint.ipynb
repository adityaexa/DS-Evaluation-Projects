Removing outliers

```python
z=np.abs(zscore(ds))
print(z.shape)
ds=ds.loc[(abs(z)<3).all(axis=1)]
print(ds.shape)
```

After removal of outliers we got actul 576 rows. 236 outliers rows has been removed.

```python
ds['Education']=np.sqrt(ds['Education'])
ds['Son']=np.sqrt(ds['Son'])
ds['Pet']=np.sqrt(ds['Pet'])
ds['Son']=np.sqrt(ds['Son'])
ds['Pet']=np.sqrt(ds['Pet'])
ds['Height']=np.sqrt(ds['Height'])
ds.skew()
```

Handling skewness of data by using numpy.sqrt method

```python
#Dropping columns which is no use for models
ds.drop(['Disciplinary failure','Social smoker'],axis=1,inplace=True)
```

```python
x=ds.drop(['ID','Absenteeism time in hours','Service time','Height','Weight','Work load Average/day '],axis=1)
print(x.shape)
y=ds['Absenteeism time in hours']
print(y.shape)
```

separate input and output data for Regression models

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

Scaled input data for model

```python
def maxr2_score(rg,x,y):
    max_r_score =0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        rg.fit(x_train,y_train)
        pred=rg.predict(x_test)
        r2_scr=r2_score(y_test,pred)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print('max r2 score corrasponding to ',final_r_state,'is',max_r_score)
    return final_r_state
```

Created function for finding the max r2 score value of model

# Applying different algorithms to create model

```python
lg=LinearRegression()
r_state=maxr2_score(lg,x,y)
print('Mean r2 score for linear Regression:',cross_val_score(lg,x,y,cv=5,scoring='r2').mean())
print('stander deviation in r2 score for linear Regression',cross_val_score(lg,x,y,cv=5,scoring='r2').std())
```

```python
neighbors={'n_neighbors':range(1,20)}
knr=KNeighborsRegressor()
gknr=GridSearchCV(knr,neighbors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=11)
r_state=maxr2_score(knr,x,y)
print('Mean r2 score for KNN Regression:',cross_val_score(knr,x,y,cv=5,scoring='r2').mean())
print('stander deviation r2 score for KNN Regression:',cross_val_score(knr,x,y,cv=5,scoring='r2').std())
```

```python
lsreg=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
clf=GridSearchCV(lsreg,parameters,cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
lsreg=Lasso(alpha=0.1)
r_state=maxr2_score(lsreg,x,y)
print('Mean r2 score for Lasso Regression',cross_val_score(lsreg,x,y,cv=5,scoring='r2').mean())
print('stander deviation in r2 score for Lasso Regression',cross_val_score(lsreg,x,y,cv=5,scoring='r2').std())
```

```python
ada_reg =AdaBoostRegressor()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000],"base_estimator":[lg,lsreg,DecisionTreeRegressor()]}
clf=GridSearchCV(ada_reg,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
ada_reg=AdaBoostRegressor(base_estimator=lsreg,learning_rate=0.001,n_estimators=1000)
r_state=maxr2_score(ada_reg,x,y)
print('Mean r2 score for ada Boosting Regration:',cross_val_score(ada_reg,x,y,cv=5,scoring='r2').mean())
print('Mean r2 score for ada Boosting Regration:',cross_val_score(ada_reg,x,y,cv=5,scoring='r2').std())
```

```python
gb=GradientBoostingRegressor()
r_state=maxr2_score(gb,x,y)
print('Mean r2 score for ada Boosting Regration:',cross_val_score(gb,x,y,cv=5,scoring='r2').mean())
print('Mean r2 score for ada Boosting Regration:',cross_val_score(gb,x,y,cv=5,scoring='r2').std())
```

```python
models=['linear', 'poly', 'rbf', 'sigmoid']
```

```python
for i in models:
    svc=SVR(kernel=i)
    print('*******************************',i,'************************************************')
    print('\n')
    r_state=maxr2_score(svc,x,y)
    print('Mean r2 score for SVC:',cross_val_score(svc,x,y,cv=5).mean())
    print('stander deviation in r2 score for SVC',cross_val_score(svc,x,y,cv=5,).std())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=68,test_size=0.20)
gb.fit(x_train,y_train)
gb_pred=gb.predict(x_test)
print('RMSE is: ',np.sqrt(mean_squared_error(y_test,gb_pred)))
print('MAE is: ',mean_absolute_error(y_test,gb_pred))
print('r2_score is:',r2_score(y_test,gb_pred))
print('cross val score Mean:',cross_val_score(gb,x,y,cv=5,scoring='r2').mean())
print('cross val score STD:',cross_val_score(gb,x,y,cv=5,scoring='r2').std())
print('\n')
print(gb_pred)
print(y_test)
```
