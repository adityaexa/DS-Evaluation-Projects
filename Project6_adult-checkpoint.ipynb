# Handling data outliers,Correlation columns and skewness

```python
for i in ds_new.columns:
    plt.boxplot(ds_new[i])
    print(i)
    print('\n')
    plt.show()
```

After using boxplot, many outliers have been found in columns.

```python
for i in ds_new.columns:
    print(i)
    sns.distplot(ds_new[i],bins=10)
    print('\n')
    plt.show()
```

Checking the distribution of columns data, we found there is uneven skewness of diffrent columns like:

Removing outliers

```python
z=np.abs(zscore(ds_new))
print(z.shape)
ds_new=ds_new.loc[(abs(z)<3).all(axis=1)]
print(ds_new.shape)
```

After removal of outliers we got actul 27418 rows. 

Around 5,143 outliers data has deleted

```python
ds_new.corr()
```

Using heatmap to check correlated columns. We have found some of the columns are highly correlated positive and negative columns(4,10,5,7,14).

```python
x=ds_new.drop([4,10,5,7,14],axis=1)
print(x.shape)
y=ds_new[14]
print(y.shape)
```

separate input and output data for classification models

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

Scaled input data for model

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=55,test_size=0.20)
```

separated data in trained test for model

```python
KNN=KNeighborsClassifier(n_neighbors=6)
SV=SVC()
LR=LogisticRegression()
DT=DecisionTreeClassifier(random_state=6)
GNB=GaussianNB()
```

Declared algorithms in variables 

```python
models=[]
models.append(('KNeighborsClassifier',KNN))
models.append(('SVC',SV))
models.append(('LogisticRegression',LR))
models.append(('DecisionTreeClassifier',DT))
models.append(('GaussianNB',GNB))
```

append algorithmsin in list

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
for name,model in models:
    print('**********************',name,'**********************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print('Accuracy_score =',AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='accuracy').mean()
    print('Cross_Val_Score =',sc)
    cvs.append(sc*100)
    print('\n')
    false_positive_rate,true_positive_rate,thresholds=roc_curve(y_test,pre)
    roc_auc=auc(false_positive_rate,true_positive_rate)
    print('roc_auc_score= ',roc_auc)
    rocscore.append(roc_auc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pre))
    print('\n')
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=False))
    plt.subplot(912)
    plt.title(name)
    plt.plot(false_positive_rate,true_positive_rate,label='AUC=%0.2f'% roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('Trur Positive Rate')
    plt.xlabel('False Positive Rate')
    print('\n\n')
```

Created anonymous program for flowing actions:

1. Run the algorithm
2. To check accuracy score of algorithms
3. To check Cross Val Score of algorithms
4. Find out the roc auc score
5. Find out the classification report
6. Visualize confusion matrix of each algorithm
6. Visualize auc curve of each algorithm

```python
modeltab=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_val_score':cvs,'Roc_cuc_curve':rocscore})
modeltab
```
